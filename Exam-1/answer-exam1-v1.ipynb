```python
allnames = ["mass_mmc","mass_trans_met_lep","mass_vis","pt_h","deltaeta_jet_jet","mass_jet_jet","prodeta_jet_jet","deltar_tau_lep","pt_tot","sum_pt","pt_ratio_lep_tau","met_phi_cen","lep_eta_cen","tau_pt","tau_eta","tau_phi","lep_pt", "lep_eta","lep_phi","met","met_phi","met_sumet","jet_num","jet_lead_pt","jet_lead_eta","jet_lead_phi","jet_sublead_pt","jet_sublead_eta","jet_sublead_phi","jet_all_pt","weight","label"]

rawname = ["tau_pt","tau_eta","tau_phi","lep_pt", "lep_eta","lep_phi","met","met_phi","met_sumet","jet_num","jet_lead_pt","jet_lead_eta","jet_lead_phi","jet_sublead_pt","jet_sublead_eta","jet_sublead_phi","jet_all_pt"]

featname = ["mass_mmc","mass_trans_met_lep","mass_vis","pt_h","deltaeta_jet_jet","mass_jet_jet","prodeta_jet_jet","deltar_tau_lep","pt_tot","sum_pt","pt_ratio_lep_tau","met_phi_cen","lep_eta_cen"]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(filename, names=allnames)
df['label'] = (df['label'] !='b').astype(int)
df
```

```python
df_sig=df[df.label==1]
df_bkg=df[df.label==0]
```

```python
for var in allnames[:30]:
    print var
    plt.figure()
    plt.hist(df_sig[var],bins=100,histtype="step", color="red",label="background",stacked=True)
    plt.hist(df_bkg[var],bins=100,histtype="step", color="blue", label="signal",stacked=True)
    plt.legend(loc='upper right')
    plt.show()
```

```python
###################################
###### Imports/Prepping Data ######
###################################

import numpy as np
from sklearn.metrics import roc_curve, auc
from sklearn import linear_model
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

N_Train = 20000

### Raw Data ###

Train_raw=df[:N_Train]
Test_raw=df[N_Train:]

X_RTrain=Train_raw[allnames[14:30]]
y_RTrain=Train_raw["label"]

X_RTest=Test_raw[allnames[14:30]]
y_RTest=Test_raw["label"]

### Feature Data ###

Train_feat=df[:N_Train]
Test_feat=df[N_Train:]

X_FTrain=Train_feat[allnames[:14]]
y_FTrain=Train_feat["label"]

X_FTest=Test_feat[allnames[:14]]
y_FTest=Test_feat["label"]

### Raw+Feature Data ###

Train_all=df[:N_Train]
Test_all=df[N_Train:]

X_ATrain=Train_all[allnames[:30]]
y_ATrain=Train_all["label"]

X_ATest=Test_all[allnames[:30]]
y_ATest=Test_all["label"]

```

```python
#############################################
###### Ordinary Least Squares Analysis ######
#############################################

lreg = linear_model.LinearRegression() 

### raw analysis

lreg.fit(X_RTrain, y_RTrain)
fpr_lreg_raw, tpr_lreg_raw, _ = roc_curve(y_RTest, lreg.decision_function(X_RTest))
roc_auc_lreg_raw = auc(fpr_lreg_raw, tpr_lreg_raw)

### feature anaylsis

lreg.fit(X_FTrain, y_FTrain)
fpr_lreg_feat, tpr_lreg_feat, _ = roc_curve(y_FTest, lreg.decision_function(X_FTest))
roc_auc_lreg_feat = auc(fpr_lreg_feat, tpr_lreg_feat)

### raw+feature analysis

lreg.fit(X_ATrain, y_ATrain)
fpr_lreg_all, tpr_lreg_all, _ = roc_curve(y_ATest, lreg.decision_function(X_ATest))
roc_auc_lreg_all = auc(fpr_lreg_all, tpr_lreg_all)


##########################################
###### Logistic Regression Analysis ######
##########################################

logreg = LogisticRegression()

### raw analysis

logreg.fit(X_RTrain, y_RTrain)
fpr_logreg_raw, tpr_logreg_raw, _ = roc_curve(y_RTest, logreg.decision_function(X_RTest))
roc_auc_logreg_raw = auc(fpr_logreg_raw, tpr_logreg_raw)

### featured analysis

logreg.fit(X_FTrain, y_FTrain)
fpr_logreg_feat, tpr_logreg_feat, _ = roc_curve(y_FTest, logreg.decision_function(X_FTest))
roc_auc_logreg_feat = auc(fpr_logreg_feat, tpr_logreg_feat)

### raw+feature analysis 

logreg.fit(X_ATrain, y_ATrain)
fpr_logreg_all, tpr_logreg_all, _ = roc_curve(y_ATest, logreg.decision_function(X_ATest))
roc_auc_logreg_all = auc(fpr_logreg_all, tpr_logreg_all)


########################################################
###### Stochastic Gradient Descent - SGD Analysis ######
########################################################

sgdc = SGDClassifier(shuffle=True,loss="hinge")

### raw analysis

sgdc.fit(X_RTrain, y_RTrain)
fpr_sgdc_raw, tpr_sgdc_raw, _ = roc_curve(y_RTest, sgdc.decision_function(X_RTest))
roc_auc_sgdc_raw = auc(fpr_sgdc_raw, tpr_sgdc_raw)

### featured analysis

sgdc.fit(X_FTrain, y_FTrain)
fpr_sgdc_feat, tpr_sgdc_feat, _ = roc_curve(y_FTest, sgdc.decision_function(X_FTest))
roc_auc_sgdc_feat = auc(fpr_sgdc_feat, tpr_sgdc_feat)

### raw+feature analysis 

sgdc.fit(X_ATrain, y_ATrain)
fpr_sgdc_all, tpr_sgdc_all, _ = roc_curve(y_ATest, sgdc.decision_function(X_ATest))
roc_auc_sgdc_all = auc(fpr_sgdc_all, tpr_sgdc_all)


```

```python
#################################################
###### Graphing ROC / Comparisons Analysis ######
#################################################

plt.figure("Comparing Classifiers", [14,14])

plt.subplot(3, 1, 1)
plt.title("Raw Analysis")
plt.plot(fpr_lreg_raw,tpr_lreg_raw,color="red", label="Ordinary Least Squares ROC (area = %0.2f)" % roc_auc_lreg_raw)
plt.plot(fpr_logreg_raw, tpr_logreg_raw, color="green", label="Logistic Regression ROC (area = %0.2f)" % roc_auc_logreg_raw)
plt.plot(fpr_sgdc_raw, tpr_sgdc_raw, color="blue", label="Stochastic Gradient Descent ROC (area = %0.2f)" % roc_auc_sgdc_raw)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.subplot(3, 1, 2)
plt.title("Features Analysis")
plt.plot(fpr_lreg_feat,tpr_lreg_feat,color="red", label="Ordinary Least Squares ROC (area = %0.2f)" % roc_auc_lreg_feat)
plt.plot(fpr_logreg_feat, tpr_logreg_feat, color="green", label="Logistic Regression ROC (area = %0.2f)" % roc_auc_logreg_feat)
plt.plot(fpr_sgdc_feat, tpr_sgdc_feat, color="blue", label="Stochastic Gradient Descent ROC (area = %0.2f)" % roc_auc_sgdc_feat)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.subplot(3, 1, 3)
plt.title("Raw+Features Analysis")
plt.plot(fpr_lreg_all,tpr_lreg_all,color="red", label="Ordinary Least Squares ROC (area = %0.2f)" % roc_auc_lreg_all)
plt.plot(fpr_logreg_all, tpr_logreg_all, color="green", label="Logistic Regression ROC (area = %0.2f)" % roc_auc_logreg_all)
plt.plot(fpr_sgdc_all, tpr_sgdc_all, color="blue", label="Stochastic Gradient Descent ROC (area = %0.2f)" % roc_auc_sgdc_all)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.show()
```
